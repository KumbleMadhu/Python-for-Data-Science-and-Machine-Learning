**Data visualizations!**

As we know, the target label will not be there in k-Means clustering type of problems in real world. But since its given in this particular Data set, we can make use of it to draw inference from the graphs.

```python
sns.set_style('whitegrid')
sns.lmplot('Room.Board','Grad.Rate',data=college_data, hue='Private',
           palette='coolwarm',height=6,aspect=1,fit_reg=False)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = 'Outstate', y = 'F.Undergrad', data = college_data, hue= 'Private')
```

```python
sns.set_style('darkgrid')
plot = sns.FacetGrid(college_data, hue= 'Private', palette= 'coolwarm', height= 6, aspect= 2)
plot = plot.map(plt.hist, 'Outstate', bins= 30, alpha= 0.7)
```

```python
sns.set_style('darkgrid')
plot = sns.FacetGrid(college_data, hue= 'Private', palette= 'coolwarm', height= 6, aspect= 2)
plot = plot.map(plt.hist, 'Grad.Rate', bins= 30, alpha= 0.7)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2, verbose=2)
```

```python
km.fit(college_data.drop('Private', axis = 1))
```

```python
km.cluster_centers_
```

```python
km.labels_
```

**So here, 1 is for a Private university, and 0 is for a Public university.**

**Since target variable won't be usually given, drawing a confusion matrix or classification report is not possible.Hence we will avoid that step. There is no perfect way to evaluate clustering if we don't have the labels.**
